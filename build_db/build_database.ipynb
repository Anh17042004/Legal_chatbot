# Tạo enities

```python
import pandas as pd
from tqdm import tqdm

def build_entities(filtered_meta, doc_to_chunks_map):
    """Tạo entity cho mỗi văn bản pháp luật."""
    print("\n" + "=" * 60)
    print("BƯỚC 3: Tạo Entities (LegalDocument)")
    print("=" * 60)

    entities = []
    for _, row in tqdm(filtered_meta.iterrows(), total=len(filtered_meta)):
        loai_vb = row["Loại Văn Bản"]
        if pd.isna(loai_vb):
            continue

        doc_id = row.get("Số Hiệu")
        if pd.isna(doc_id):
            continue
        doc_id = str(doc_id).strip()
        if not doc_id or doc_id.lower() == "nan":
            continue

        # Ghép description từ metadata — chứa cả IssuingBody
        parts = []
        parts.append(f"{loai_vb}")  # loại văn bản
        if pd.notna(row.get("Số Hiệu")):
            parts.append(f"{row['Số Hiệu']}")
        if pd.notna(row.get("Chủ Đề")):
            parts.append(f"{row['Chủ Đề']}")
        if pd.notna(row.get("Nơi Ban Hành")):
            parts.append(f"Được ban hành bởi: {row['Nơi Ban Hành']}")
        if pd.notna(row.get("Hiệu lực")):
            parts.append(f"{row['Hiệu lực']}")

        description = ". ".join(parts)

        chunk_ids = doc_to_chunks_map.get(doc_id, [])
        source_id = "<SEP>".join(chunk_ids) if chunk_ids else doc_id

        entities.append(
            {
                "entity_name": row['Số Hiệu'],
                "entity_type": "LegalDocument", 
                "description": description,
                "source_id": source_id,
                "file_path": str(row['file_path']),
            }
        )

    print(f"  → Tạo {len(entities)} entities")
    return entities
```

```python
REL_MAP = {
    "Văn bản hướng dẫn": {
        "kw": "hướng dẫn, quy định chi tiết",
        "w": 0.95
    },
    "Văn bản căn cứ": {
        "kw": "căn cứ, cơ sở pháp lý",
        "w": 1.0
    },
    "Văn bản được hướng dẫn": {
        "kw": "được hướng dẫn, được quy định chi tiết",
        "w": 0.95
    },
    "Văn bản quy định hết hiệu lực một phần": {
        "kw": "bãi bỏ, hết hiệu lực một phần",
        "w": 0.85
    },
    "Văn bản hết hiệu lực một phần": {
        "kw": "hết hiệu lực một phần",
        "w": 0.8
    },
    "Văn bản được hợp nhất": {
        "kw": "được hợp nhất",
        "w": 0.75
    },
    "Văn bản sửa đổi, bổ sung": {
        "kw": "sửa đổi, bổ sung",
        "w": 0.9
    },
    "Văn bản dẫn chiếu": {
        "kw": "dẫn chiếu, tham chiếu, viện dẫn",
        "w": 0.7
    },
    "Văn bản bị sửa đổi, bổ sung": {
        "kw": "bị sửa đổi, bổ sung",
        "w": 0.9
    },
    "Văn bản hết hiệu lực": {
        "kw": "hết hiệu lực, bị bãi bỏ",
        "w": 0.85
    },
    "Văn bản thay thế": {
        "kw": "thay thế",
        "w": 0.9
    },
    "Văn bản được hợp nhất (sửa đổi)": {
        "kw": "được hợp nhất, sửa đổi",
        "w": 0.8
    },
}

def build_relationships(df_rels, doc_to_chunks_map):
    """Tạo relationships giữa các văn bản từ bảng rels."""
    print("\n" + "=" * 60)
    print("BƯỚC 4: Tạo Relationships từ bảng rels")
    print("=" * 60)

    relationships = []
    skipped = 0

    for _, row in tqdm(df_rels.iterrows(), total=len(df_rels)):
        rel_type_vi = str(row["mối quan hệ"]).strip()

        rel_info = REL_MAP.get(rel_type_vi)
        if not rel_info:
            skipped += 1
            continue

        doc_id = str(row["số ký hiệu"]).strip()
        tgt_id = str(row["số ký hiệu văn bản liên quan"]).strip()

        if not doc_id or not tgt_id or doc_id.lower() == "nan" or tgt_id.lower() == "nan" or doc_id == tgt_id:
            skipped += 1
            continue

        chunk_ids = doc_to_chunks_map.get(doc_id, [])
        source_id = "<SEP>".join(chunk_ids) if chunk_ids else doc_id

        relationships.append(
            {
                "src_id": doc_id,
                "tgt_id": tgt_id,
                "description": f"{doc_id} {rel_info['kw'].split(',')[0].strip()} {tgt_id}",
                "keywords": rel_info["kw"],
                "weight": rel_info["w"],
                "source_id": source_id,
                "file_path": doc_id,
            }
        )

    print(f"  → Tạo {len(relationships)} relationships (bỏ {skipped} dòng)")
    return relationships
```

# Build Chunks

```python
import re
from bs4 import BeautifulSoup
import json

from pyvi import ViTokenizer
from langchain_experimental.text_splitter import SemanticChunker
from langchain_huggingface import HuggingFaceEmbeddings

model_kwargs = {"device": "cuda"}
encode_kwargs = {"normalize_embeddings": True}
embeddings_Model = HuggingFaceEmbeddings(
    model_name="huyydangg/DEk21_hcmute_embedding",
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)


chunker = SemanticChunker(
    embeddings = embeddings_Model,
    buffer_size = 5,
    breakpoint_threshold_type = "gradient",
    breakpoint_threshold_amount = 0.75,
    sentence_split_regex = r"(?<=[.?!])\s+(?!\d+\.\s)",
    min_chunk_size = 512
)

def count_tokens(text):
    return len(ViTokenizer.tokenize(text).split())


def split_article_body(full_content):
    clean_content = full_content.strip()
    if not clean_content:
        return [""]

    token_len = count_tokens(clean_content)
    if token_len <= 1000:
        return [clean_content]

    try:
        semantic_parts = [
            part.strip()
            for part in chunker.split_text(clean_content)
            if part and part.strip()
        ]
    except Exception:
        return [clean_content]

    return semantic_parts or [clean_content]

def parse_to_chunks(content, so_ky_hieu):
    if not content or pd.isna(content):
        return []

    # soup = BeautifulSoup(html_content, "html.parser")
    # text = soup.get_text(separator="\n")
    lines = [line.strip() for line in content.split("\n") if line.strip()]

    current_chuong = ""
    current_muc = ""
    current_tieu_muc = ""
    current_dieu = ""
    current_content = []
    chunks = []

    chuong_pattern = re.compile(r"^(Chương|CHƯƠNG)\s+([IVXLCDM]+|\d+)(\.|:)?")
    muc_pattern = re.compile(r"^(Mục|MỤC)\s+([IVXLCDM]+|\d+)(\.|:)?")
    tieu_muc_pattern = re.compile(r"^(Tiểu mục|TIỂU MỤC)\s+([IVXLCDM]+|\d+)(\.|:)?")
    dieu_pattern = re.compile(r"^Điều\s+(\d+)\.")

    def save_current_dieu():
        nonlocal current_dieu, current_content
        if not current_dieu:
            return

        full_content = "\n".join(current_content).strip()

        prefix = f"[{so_ky_hieu}]"
        if current_chuong:
            prefix += f" [{current_chuong.strip()}]"
        if current_muc:
            prefix += f" [{current_muc.strip()}]"
        if current_tieu_muc:
            prefix += f" [{current_tieu_muc.strip()}]"

        dieu_match = re.search(r"Điều\s+(\d+[a-z]?)", current_dieu)
        dieu_num = dieu_match.group(1) if dieu_match else "0"
        chunk_order_index = int(re.sub(r"[a-z]", "", dieu_num))
        semantic_parts = split_article_body(full_content)
        base_source_id = f"chunk_{so_ky_hieu}_{dieu_num}"

        for part_index, part_content in enumerate(semantic_parts, start=1):
            source_id = base_source_id
            if len(semantic_parts) > 1:
                source_id = f"{base_source_id}_s{part_index}"

            chunk_text = f"{prefix} {current_dieu}\n{part_content}".strip()
            chunks.append(
                {
                    "content": chunk_text,
                    "source_id": source_id,
                    "chunk_order_index": chunk_order_index,
                    "file_path": so_ky_hieu,
                }
            )

    for line in lines:
        chuong_match = chuong_pattern.match(line)
        if chuong_match:
            current_chuong = line
            current_muc = ""
            current_tieu_muc = ""
            continue

        muc_match = muc_pattern.match(line)
        if muc_match:
            current_muc = line
            current_tieu_muc = ""
            continue

        tieu_muc_match = tieu_muc_pattern.match(line)
        if tieu_muc_match:
            current_tieu_muc = line
            continue

        dieu_match = dieu_pattern.match(line)
        if dieu_match:
            save_current_dieu()
            current_dieu = line
            current_content = []
        else:
            if current_dieu:
                current_content.append(line)

    save_current_dieu()
    return chunks


def build_chunks(df_content):
    """Parse HTML của tất cả văn bản thành chunks pháp lý + semantic sub-chunks."""
    print("\n" + "=" * 60)
    print("BƯỚC 2: Parse HTML -> Chunks pháp lý + semantic sub-chunks")
    print("=" * 60)

    all_chunks = []
    doc_to_chunks_map = {}
    docs_with_chunks = 0
    docs_without_chunks = 0

    for _, row in tqdm(df_content.iterrows(), total=len(df_content)):
        doc_id = str(row["Số Ký Hiệu"])
        if not doc_id:
            continue

        chunks = parse_to_chunks(row["Content"], doc_id)
        if chunks:
            all_chunks.extend(chunks)
            doc_to_chunks_map[doc_id] = [c["source_id"] for c in chunks]
            docs_with_chunks += 1
        else:
            docs_without_chunks += 1

    print(f"  Tạo {len(all_chunks)} chunks từ {docs_with_chunks} VB")
    print(
        f"  {docs_without_chunks} VB không parse được chunks (HTML rỗng hoặc không có Điều)"
    )
    return all_chunks, doc_to_chunks_map
```

```python
import asyncio
import json
import os
import random
import os
from lightrag import LightRAG, QueryParam
from lightrag.llm.ollama import ollama_model_complete
from lightrag.operate import merge_nodes_and_edges
from dataclasses import asdict
from lightrag.utils import EmbeddingFunc
import numpy as np

# --- CẤU HÌNH HUGGINGFACE EMBEDDING (TIẾNG VIỆT) ---
# Yêu cầu cài đặt: pip install sentence-transformers langchain-community
from langchain_community.embeddings import HuggingFaceEmbeddings

print("Đang khởi tạo mô hình HuggingFace Embedding...")
hf_model = HuggingFaceEmbeddings(
    model_name="huyydangg/DEk21_hcmute_embedding",
    model_kwargs={"device": "cuda"}, # 👉 ĐỔI THÀNH "cuda" nếu bạn có card đồ họa NVIDIA
    encode_kwargs={"normalize_embeddings": True}
)

# --- CẤU HÌNH CLOUD DATABASES (ZILLIZ MILVUS & NEO4J) ---
def require_env(name: str) -> str:
    value = os.getenv(name)
    if not value:
        raise ValueError(f"Thiếu biến môi trường bắt buộc: {name}")
    return value

os.environ["MILVUS_URI"] = "https://in03-849a81265e273a4.serverless.aws-eu-central-1.cloud.zilliz.com" 
os.environ["MILVUS_TOKEN"] = "339079fe154fdcf363221161288bfbd58a093cb62ba0ce4bd1c572db3e89b063367fbc5dd89a70f2200e4b5289c2860e41134c0b" 
os.environ["MILVUS_DB_NAME"] = ""

# Neo4j Cloud
os.environ["NEO4J_URI"] = "neo4j+s://dc053a4e.databases.neo4j.io"
os.environ["NEO4J_USERNAME"] = "dc053a4e"
os.environ["NEO4J_PASSWORD"] = "DlzLxZsP7T8r_tAIWcv5NBp3Qn3Q4KpEQZ5QX-IN57I"

# Wrapper bọc Langchain lại để trả về chuẩn Numpy Array mà LightRAG quy định
async def custom_hf_embedding(texts: list[str]) -> np.ndarray:
    results = hf_model.embed_documents(texts)
    return np.array(results)


# --- CƠ CHẾ XOAY VÒNG API KEY (API KEY ROTATION) ---
# Bạn điền danh sách các API Keys của Ollama Cloud vào biến môi trường OLLAMA_API_KEYS, phân tách bằng dấu phẩy.
API_KEYS = [
     "d4ba96ab5ddf4c12b43f9fcca0384f5c.h0gastU_cO5XpBu_s03VddNX",
    "6cacdbee3c57494886be11756a92efd5.IclooWVhq0nHjTbslQ5uVoaa",
    "76a3a3ae007e450bba6492fb27430533.ip9J4qQ1i2LTpdmQXu2UILG2",
    "6a86fdabfa40402dbd0cfce7cbd0da08.KgCz2YzvV9lrYLNKHi5XEzq5"
]
if not API_KEYS:
    raise ValueError("Thiếu biến môi trường bắt buộc: OLLAMA_API_KEYS")
current_key_idx = 0

async def rotating_ollama_model_complete(prompt, system_prompt=None, history_messages=[], **kwargs):
    global current_key_idx

    # Cho phép cố gắng xoay vòng gấp đôi số lượng key để đảm bảo an toàn
    max_retries = len(API_KEYS) * 2

    for attempt in range(max_retries):
        try:
            # Gán key hiện hành vào tham số gọi API
            kwargs["api_key"] = API_KEYS[current_key_idx]

            # Gọi hàm gốc của LightRAG
            result = await ollama_model_complete(
                prompt,
                system_prompt=system_prompt,
                history_messages=history_messages,
                **kwargs
            )

            # --- TẠO QUÃNG NGHỈ (THROTTLE) ĐỂ TRÁNH BỊ RATE LIMIT ---
            # Random nghỉ từ 1 đến 3 giây giữa mỗi Request thành công
            await asyncio.sleep(random.uniform(10.0, 15.0))

            return result

        except Exception as e:
            error_str = str(e).lower()
            # Nếu bắt trúng các lỗi kiểu Rate Limit, 429, Quota Exceeded...
            if "limit" in error_str or "429" in error_str or "quota" in error_str or "too many requests" in error_str:
                print(f"[CẢNH BÁO] API Key {API_KEYS[current_key_idx][:10]}... bị sập (Rate Limit/Quota). Đang Switch sang Key mới...")
                # Đi tới Key tiếp theo trong mảng, nếu hết mảng thì quay về số 0
                current_key_idx = (current_key_idx + 1) % len(API_KEYS)
            else:
                # Nếu không phải lỗi Rate limit (ví dụ mất mạng) -> quăng lỗi liền để Fix
                raise e

    raise Exception("Thất bại! Đã thử đổi toàn bộ API Keys nhưng tất cả đều bị Limit/Quota.")

# Import các hàm khởi tạo embedding và graph tuỳ thuộc vào setup của dự án bạn
# Ví dụ: từ utils của LightRAG hoặc tự định nghĩa HuggingFace

async def build_macro_layer(rag, custom_entities_dict, custom_relations_dict):
    print("\n--- GIAI ĐOẠN 2: NẠP TẦNG VĨ MÔ (MACRO) ---")
    print("1. Nạp Entities/Relations Vĩ Mô vào VectorDB...")
    if custom_entities_dict:
        await rag.entities_vdb.upsert(custom_entities_dict)
    if custom_relations_dict:
        await rag.relationships_vdb.upsert(custom_relations_dict)

    print("2. Nạp Node/Edge Vĩ mô vào GraphDB (Neo4j/NetworkX)...")
    for ent in custom_entities_dict.values():
        await rag.chunk_entity_relation_graph.upsert_node(ent["entity_name"], ent)

    for rel in custom_relations_dict.values():
        await rag.chunk_entity_relation_graph.upsert_edge(rel["src_id"], rel["tgt_id"], rel)
    print("-> Hoàn tất nạp tầng vĩ mô.")

async def build_micro_layer(rag, custom_chunks_dict):
    print("\n--- GIAI ĐOẠN 3: NẠP TẦNG VI MÔ (MICRO) ---")
    print("3. Nạp Data Nguyên Bản vào Local Text Storage...")
    if custom_chunks_dict:
        await rag.text_chunks.upsert(custom_chunks_dict)

    print("4. Nạp Embeddings Data Nguyên Bản vào VectorDB...")
    if custom_chunks_dict:
        await rag.chunks_vdb.upsert(custom_chunks_dict)

    print("5. [TỐN TIỀN/TỐN THỜI GIAN NHẤT] Kích hoạt LLM Extraction theo lô BATCH...")
    # Hàm bí mật ép LLM tuân thủ Chunk_id của bạn trỏ vào Micro Entity.
    if custom_chunks_dict:
        import itertools
        BATCH_SIZE = 20 # Cứ gặm xong 20 Chunk thì đẩy thẳng lên Mây!

        # Cắt từ điển lớn thành danh sách chứa các lô nhỏ
        items = list(custom_chunks_dict.items())
        batches = [dict(items[i:i + BATCH_SIZE]) for i in range(0, len(items), BATCH_SIZE)]

        for idx, batch_dict in enumerate(batches):
            print(f"\n🚀 Đang cày Cuốc Lô thứ {idx + 1}/{len(batches)} (Mỗi lô {BATCH_SIZE} Chunks)...")

            # CHUẨN BỊ MÔI TRƯỜNG STATUS ĐỂ NÉ TÌNH TRẠNG LỖI NoneType Status Lock
            pipeline_status = {"history_messages": [], "request_pending": False, "latest_message": ""}
            pipeline_status_lock = asyncio.Lock()

            # 1. MỒI LLM EXTRACT (Lưu tạm trên RAM)
            chunk_results = await rag._process_extract_entities(
                batch_dict,
                pipeline_status=pipeline_status,
                pipeline_status_lock=pipeline_status_lock
            )

            # 2. XẢ TỪ RAM ĐẨY LÊN CLOUD NEO4J VÀ MILVUS (Tiến trình quyết định)
            print(f"⏳ Đang Trộn và Up Lô {idx + 1} lên Cloud Neo4j/Milvus để chốt sổ...")
            await merge_nodes_and_edges(
                chunk_results=chunk_results,
                knowledge_graph_inst=rag.chunk_entity_relation_graph,
                entity_vdb=rag.entities_vdb,
                relationships_vdb=rag.relationships_vdb,
                global_config=asdict(rag),
                full_entities_storage=rag.full_entities,
                full_relations_storage=rag.full_relations,
                doc_id=f"custom_import_batch_{idx}", # Nhãn giả định theo lô
                pipeline_status=pipeline_status,
                pipeline_status_lock=pipeline_status_lock,
                llm_response_cache=rag.llm_response_cache,
                entity_chunks_storage=rag.entity_chunks,
                relation_chunks_storage=rag.relation_chunks,
                current_file_number=1,
                total_files=1,
                file_path="/kaggle/working/custom_kg_full.json"
            )

    print("-> Hoàn tất rút trích & nạp tầng vi mô lên Cloud!")
async def main():
    # 1) Đọc dữ liệu JSON đã build trước đó
    json_path = "/kaggle/working/custom_kg_full.json"
    if not os.path.exists(json_path):
        print(f"Lỗi: Không tìm thấy file {json_path}. Vui lòng chạy bước build custom_kg trước.")
        return

    print(f"Đang tải dữ liệu từ {json_path}...")
    with open(json_path, "r", encoding="utf-8") as f:
        custom_kg = json.load(f)

    # 2) Chuẩn hóa entity/relationship/chunk theo contract của LightRAG
    custom_entities_dict = {
        ent["entity_name"]: {
            **ent,
            "entity_id": ent["entity_name"],
            "content": f"{ent['entity_name']}: {ent.get('description', '')}",
            "file_path": ent.get("file_path", "unknown_source"),
        }
        for ent in custom_kg.get("entities", [])
    }

    custom_relations_dict = {
        f"{rel['src_id']}-{rel['tgt_id']}": {
            **rel,
            "content": rel.get("description", ""),
            "file_path": rel.get("file_path", "unknown_source"),
        }
        for rel in custom_kg.get("relationships", [])
    }

    # Giữ file_path để hệ thống citation/reference hoạt động đúng nguồn
    custom_chunks_dict = {
        chunk["source_id"]: {
            "content": chunk.get("content", ""),
            "source_id": chunk["source_id"],
            "file_path": chunk.get("file_path", "unknown_source"),
            "full_doc_id": chunk.get("file_path", chunk["source_id"]),
            "chunk_order_index": chunk.get("chunk_order_index", 0),
        }
        for chunk in custom_kg.get("chunks", [])
        if chunk.get("source_id")
    }

    print(f"Loaded {len(custom_entities_dict)} macro entities")
    print(f"Loaded {len(custom_relations_dict)} macro relationships")
    print(f"Loaded {len(custom_chunks_dict)} text chunks")

    # 3) Khởi tạo LightRAG
    WORKING_DIR = "./target_db"
    if not os.path.exists(WORKING_DIR):
        os.mkdir(WORKING_DIR)

    rag = LightRAG(
        working_dir=WORKING_DIR,
        llm_model_max_async=1,
        vector_storage="MilvusVectorDBStorage",
        graph_storage="Neo4JStorage",
        llm_model_func=rotating_ollama_model_complete,
        llm_model_name="gpt-oss:120b-cloud",
        embedding_func=EmbeddingFunc(
            embedding_dim=768,
            max_token_size=256,
            func=custom_hf_embedding,
        ),
    )

    print("\nĐang khởi tạo các kết nối tới Milvus, Neo4j và KVs...")
    await rag.initialize_storages()

    print("\nBắt đầu Pipeline Data Ingestion...")
    await build_macro_layer(rag, custom_entities_dict, custom_relations_dict)
    await build_micro_layer(rag, custom_chunks_dict)

    print("\n🚀 HOÀN TẤT NHÚNG DỮ LIỆU THÀNH CÔNG VÀO LIGHTRAG!")
```
